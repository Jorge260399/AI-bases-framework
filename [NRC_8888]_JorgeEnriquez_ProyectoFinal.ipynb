# **Librer√≠as**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
import time
from sklearn import svm
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sn
```

# **Cargar el dataset**

```python
data = pd.read_csv('Banana_Ripeness_Dataset.csv')

df = pd.DataFrame(data)

df.head(15)
```

# **Verificar y limpiar de datos**

```python
df = df.drop(['view', 'SampleID'], axis=1)
df.head(15)
```

# **Identificar X y Y**

```python
x= df.drop(['class'], axis=1)
y= df['class']
```

# **Ingenier√≠a de caracter√≠sticas**

### **Label Encoding**

```python
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(y)
Y = pd.DataFrame(integer_encoded, columns=["class"])
Y.head(15)
```

### **Normalizaci√≥n**

```python
scaler_mean = StandardScaler(with_mean=True, with_std=False)


scaler_minmax = RobustScaler(with_centering=False,
                             with_scaling=True,
                             quantile_range=(0, 100))
scaler_mean.fit(x)
scaler_minmax.fit(x.values)

# transform train and test sets
X_scaled = scaler_minmax.transform(scaler_mean.transform(x))
```

```python
X_scaled = pd.DataFrame(X_scaled, columns=x.columns)
```

```python
np.round(x.describe(), 1)
```

```python
np.round(X_scaled.describe(), 1)
```

# **Training and Testing**

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, Y.values.reshape(-1,1),
                                        test_size   = 0.2,
                                        random_state = 1234,
                                        shuffle      = True)

X_train.shape, X_test.shape
```

# **Grid Search**

```python
param_grid = [
    {
        "kernel": ["linear"],
		    "C": [0.0001, 0.001, 0.1, 1, 10, 100, 1000],
        "decision_function_shape": ['ovo', 'ovr']
    },
    {
        "kernel": ["poly"],
        "gamma": [1,0.1,0.01,0.001],
		    "degree": [2, 3, 4],
		    "C": [0.0001, 0.001, 0.1, 1, 10, 100, 1000],
        "decision_function_shape": ['ovo', 'ovr']
     
    },
	  {
        "kernel": ["rbf", "sigmoid"],
		    "gamma": [1,0.1,0.01,0.001],
		    "C": [0.0001, 0.001, 0.1, 1, 10, 100, 1000],
        "decision_function_shape": ['ovo', 'ovr']
    }
]
```

```python
grid = GridSearchCV(estimator=svm.SVC(), param_grid=param_grid, refit=True, verbose=2)
start = time.time()
grid.fit(X_train, y_train.ravel())
end = time.time()
```

```python
print("[INFO] grid search tard√≥ {:.2f} segundos".format(
	end - start))
print("[INFO] El mejor puntaje de grid search es: {:.2f}%".format(
	grid.best_score_ * 100))
print("[INFO] Los mejores parametros de grid search son: {}".format(
	grid.best_params_))
```

# **SVM Model**

```python
linear = svm.SVC(C=100, decision_function_shape='ovo', kernel='linear').fit(X_train, y_train.ravel())
rbf = svm.SVC(kernel='rbf', gamma=0.01, C=1000).fit(X_train, y_train.ravel())
poly = svm.SVC(decision_function_shape='ovo',kernel='poly', degree=3, C=1000).fit(X_train, y_train.ravel())
sigmoid = svm.SVC(kernel='sigmoid', gamma=0.1, C=100).fit(X_train, y_train.ravel())
```

```python
linear_pred = linear.predict(X_test)
poly_pred = poly.predict(X_test)
rbf_pred = rbf.predict(X_test)
sigmoid_pred = sigmoid.predict(X_test)
```

# **Model Evaluation**

```python
confMatrix = confusion_matrix(y_test, linear_pred)

ax = sn.heatmap(confMatrix , annot=True)

ax.set_xlabel("Predicted")
ax.set_ylabel("Actual")
plt.show()
```

```python
linear_accuracy = accuracy_score(y_test, linear_pred)
linear_precision = precision_score(y_test, linear_pred, average='weighted')
linear_recall = recall_score(y_test, linear_pred, average='weighted')
linear_f1 = f1_score(y_test, linear_pred, average='weighted')
print('Accuracy (Linear Kernel): ', "%.2f" % (linear_accuracy*100))
print('F1 (Linear Kernel): ', "%.2f" % (linear_f1*100))
print('Precision (Linear Kernel): ', "%.2f" % (linear_precision*100))
print('Recall (Linear Kernel): ', "%.2f" % (linear_recall*100))
```

```python
confMatrix = confusion_matrix(y_test, poly_pred)

ax = sn.heatmap(confMatrix , annot=True)

ax.set_xlabel("Predicted")
ax.set_ylabel("Actual")
plt.show()
```

```python
poly_accuracy = accuracy_score(y_test, poly_pred)
poly_precision = precision_score(y_test, poly_pred, average='weighted')
poly_recall = recall_score(y_test, poly_pred, average='weighted')
poly_f1 = f1_score(y_test, poly_pred, average='weighted')
print('Accuracy (Polynomial Kernel): ', "%.2f" % (poly_accuracy*100))
print('F1 (Polynomial Kernel): ', "%.2f" % (poly_f1*100))
print('Precision (Polynomial Kernel): ', "%.2f" % (poly_precision*100))
print('Recall (Polynomial Kernel): ', "%.2f" % (poly_recall*100))
```

```python
confMatrix = confusion_matrix(y_test, rbf_pred)

ax = sn.heatmap(confMatrix , annot=True)

ax.set_xlabel("Predicted")
ax.set_ylabel("Actual")
plt.show()
```

```python
rbf_accuracy = accuracy_score(y_test, rbf_pred)
rbf_precision = precision_score(y_test, rbf_pred, average='weighted')
rbf_recall = recall_score(y_test, rbf_pred, average='weighted')
rbf_f1 = f1_score(y_test, rbf_pred, average='weighted')
print('Accuracy (RBF Kernel): ', "%.2f" % (rbf_accuracy*100))
print('F1 (RBF Kernel): ', "%.2f" % (rbf_f1*100))
print('Precision (RBF Kernel): ', "%.2f" % (rbf_precision*100))
print('Recall (RBF Kernel): ', "%.2f" % (rbf_recall*100))
```

```python
confMatrix = confusion_matrix(y_test, sigmoid_pred)

ax = sn.heatmap(confMatrix , annot=True)

ax.set_xlabel("Predicted")
ax.set_ylabel("Actual")
plt.show()
```

```python
sigmoid_accuracy = accuracy_score(y_test, sigmoid_pred)
sigmoid_precision = precision_score(y_test, sigmoid_pred, average='weighted')
sigmoid_recall = recall_score(y_test, sigmoid_pred, average='weighted')
sigmoid_f1 = f1_score(y_test, sigmoid_pred, average='weighted')
print('Accuracy (Sigmoid Kernel): ', "%.2f" % (sigmoid_accuracy*100))
print('F1 (Sigmoid Kernel): ', "%.2f" % (sigmoid_f1*100))
print('Precision (Sigmoid Kernel): ', "%.2f" % (sigmoid_precision*100))
print('Recall (Sigmoid Kernel): ', "%.2f" % (sigmoid_recall*100))
```
